```python
#coperation name & job name
import requests
import bs4
import json

url = "https://www.104.com.tw/jobs/search/?keyword=Python&order=1&jobsource=2018indexpoc&ro=0"
htmlFile = requests.get(url)
objSoup = bs4.BeautifulSoup(htmlFile.text, 'lxml')
jobs = objSoup.find_all('article', class_='js-job-item')
job_list = []
for job in jobs:
    cust_name = job.get('data-cust-name')
    print("公司名稱：", cust_name)
    job_name = job.get('data-job-name')
    print("職務名稱：", job_name)
    d = [('公司名稱', cust_name), ('職務名稱', job_name)]
    j_dict = dict(d)    #字典
    job_list.append(j_dict)
    
myjob = {'Job':job_list}

fn = 'job_name.json'
with open(fn, 'w') as fnObj:
    json.dump(myjob, fnObj, indent=2, ensure_ascii=False)
```

```python
#職業分類 #會變成一個字元一格
import requests
import bs4
import csv

#url = "https://occupation.taiwanjobs.gov.tw/News2.aspx?n=7&sms=8450"


def job_class(url):
    htmlFile = requests.get(url)
    objSoup = bs4.BeautifulSoup(htmlFile.text, 'lxml')
    contents = objSoup.find_all('div', class_='caption')
    title = objSoup.find('title')
    with open('classification.csv', 'w', newline='') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow(title.text)
        print(title.text)
        for content in contents:
            csvWriter.writerow(content.text)
            print(content.text)
    #csvFile.close()
    
url_H = 'https://occupation.taiwanjobs.gov.tw/News2.aspx?n='
url_T = '&sms=8450'
for page in range(4,27):
    url = url_H + str(page) + url_T
    job_class(url)
    
```

```python
#職業分類 #\完工/
import requests
import bs4
import csv

#url = "https://occupation.taiwanjobs.gov.tw/News2.aspx?n=7&sms=8450"

list_try = []
   
url_H = 'https://occupation.taiwanjobs.gov.tw/News2.aspx?n='
url_T = '&sms=8450'
for page in range(4,27):
    url = url_H + str(page) + url_T
    htmlFile = requests.get(url)
    objSoup = bs4.BeautifulSoup(htmlFile.text, 'lxml')
    contents = objSoup.find_all('div', class_='caption')
    title = objSoup.find('title')
    title_Y = ""         #抓取標題後半段想要的字們
    for i in range(17, len(title.text)):  
        title_Y += title.text[i]
    #print(title_Y) #test
    #list_try.append([title.text])
    list_try.append([title_Y])
    print("\n")
    print(title_Y)
    for content in contents:
            list_try.append([content.text])
            print(content.text)
    list_try.append([" "])
    
with open('classification.csv','w',newline='') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerows(list_try)
    
```
